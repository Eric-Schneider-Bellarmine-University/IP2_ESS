```python
import pandas
import numpy
import datetime
```

```python
#Step 1
df=pandas.read_csv('weather_2012.csv')
```

```python
#step 2
df.head()
```

```python
#Step 3
origin_columns=df.columns
origin_columns
```

```python
#Step 4
new_columns=['date_time', 'temp_c', 'dew_point', 'relative_humidity', 'wind_speed', 'visibility', 'pressure', 'weather']
df.columns=new_columns
print(df.columns)
print(origin_columns)
```

```python
#Step 5
df.describe()
```

```python
#Step 6
df['temp_f']=(df.temp_c*1.8)+32
df['temp_f']
```

```python
#Step 7
df=df[['date_time', 'temp_f', 'temp_c', 'dew_point', 'relative_humidity', 'wind_speed', 'visibility', 'pressure', 'weather']]
df.head()
```

```python
#Step 8
df.dtypes
```

```python
#Step 9
df.date_time=pandas.to_datetime(df.date_time)
df.dtypes
```

```python
#Step 10
month=df.date_time.astype(str)
month=month.str[5:7]
month=month.astype(int)
month
```

```python
monthly_avg=pandas.DataFrame((df.temp_f.groupby(month).mean(numeric_only=True)))
monthly_avg.columns
```

```python
#step 11
monthly_avg['month']=monthly_avg.index
monthly_avg
```

```python
#Step 12
monthly_avg=monthly_avg[['month','temp_f']]
monthly_avg
```

```python
#Step 13
monthly_avg.index.name=None
monthly_avg
```

```python
#Step 14
def set_month(n):
    months={1:'January',
           2:'Febuary',
           3:'March',
           4:'April',
           5:'May',
           6:'June',
           7:'July',
           8:'August',
           9:'September',
           10:'October',
           11:'November',
           12:'December'}
    return months[n]
monthly_avg['month']=list(map(lambda x:set_month(x),monthly_avg['month']))
monthly_avg
```

```python
#Step 15
hour=df.date_time.astype(str)
hour=hour.str[11:13]
hour=hour.astype(int)
hour_avg=(df.temp_f.groupby(hour).mean(numeric_only=True))
hour_avg
```

```python
#Step 16
hour_avg.index.name='Hour'
hour_avg
```
