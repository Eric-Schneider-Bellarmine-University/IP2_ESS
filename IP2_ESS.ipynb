```python
import pandas
import numpy
import datetime
```

```python
#Step 1
df=pandas.read_csv('weather_2012.csv')
```

```python
#step 2
df.head()
```

```python
#Step 3
origin_columns=df.columns
origin_columns
```

```python
#Step 4
new_columns=['date_time', 'temp_c', 'dew_point', 'relative_humidity', 'wind_speed', 'visibility', 'pressure', 'weather']
df.columns=new_columns
print(df.columns)
print(origin_columns)
```

```python
#Step 5
df.describe()
```

```python
#Step 6
df['temp_f']=(df.temp_c*1.8)+32
df['temp_f']
```

```python
#Step 7
df=df[['date_time', 'temp_f', 'temp_c', 'dew_point', 'relative_humidity', 'wind_speed', 'visibility', 'pressure', 'weather']]
df.head()
```

```python
#Step 8
df.dtypes
```

```python
#Step 9
df.date_time=pandas.to_datetime(df.date_time)
df.dtypes
```

```python
#Step 10
month=df.date_time.astype(str)
month=month.str[5:7]
month=month.astype(int)
month
```

```python
monthly_avg=pandas.DataFrame((df.temp_f.groupby(month).mean(numeric_only=True)))
monthly_avg.columns
```

```python
#step 11
monthly_avg['month']=monthly_avg.index
monthly_avg
```

```python
#Step 12
monthly_avg=monthly_avg[['month','temp_f']]
monthly_avg
```

```python
#Step 13
monthly_avg.index.name=None
monthly_avg
```

```python
#Step 14
def set_month(n):
    months={1:'January',
           2:'Febuary',
           3:'March',
           4:'April',
           5:'May',
           6:'June',
           7:'July',
           8:'August',
           9:'September',
           10:'October',
           11:'November',
           12:'December'}
    return months[n]
monthly_avg['month']=list(map(lambda x:set_month(x),monthly_avg['month']))
monthly_avg
```

```python
#Step 15
hour=df.date_time.astype(str)
hour=hour.str[11:13]
hour=hour.astype(int)
hourly_avg=(df.temp_f.groupby(hour).mean(numeric_only=True))
hourly_avg
```

```python
#Step 16
hourly_avg.index.name='Hour'
hourly_avg
```

```python
#Step 17
def set_hour(n):
    hours={0:'12:00 AM',
          1:'1:00 AM',
          2:'2:00 AM',
          3:'3:00 AM',
          4:'4:00 AM',
          5:'5:00 AM',
          6:'6:00 AM',
          7:'7:00 AM',
          8:'8:00 AM',
          9:'9:00 AM',
          10:'10:00 AM',
          11:'11:00 AM',
          12:'12:00 PM',
          13:'1:00 PM',
          14:'2:00 PM',
          15:'3:00 PM',
          16:'4:00 PM',
          17:'5:00 PM',
          18:'6:00 PM',
          19:'7:00 PM',
          20:'8:00 PM',
          21:'9:00 PM',
          22:'10:00 PM',
          23:'11:00 PM'}
    return hours[n]
hourly_avg.index=list(map(lambda x: set_hour(x),hourly_avg.index))
hourly_avg.index.name='Hour'
hourly_avg
```

```python
#Step 18
df.weather.groupby(df.weather).count()
```

```python
#Step 19
weather=list(df.weather)
weather_list=[i.split(',') for i in weather]
weathers=list(set(sum(weather_list,[])))
print(weathers)
```

```python
# Step 20
sorted_weather=sorted(weathers)
print(sorted_weather)
```

```python
#Step 21
original_columns=list(origin_columns)
original_columns.insert(1,'Temp (F)')
df.columns=original_columns
df.columns
```

```python
#Step 22
df.to_csv('weather_2021_new.csv',index=False)
```
